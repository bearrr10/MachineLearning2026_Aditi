##### The Gapminder dataset provides population data from 1952 to 2007 (at 5 year Intervals) for several countries around the world. Compare the populations of the European countries France, United Kingdom, Italy, Germany and Spain over this period using a line chart. Make appropriate modifications to the chart title, axis titles, legend, figure size, font size, colors etc. to make the chart readable and visually appealing- Line Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:/ml/MachineLearning2026_Aditi/Datasets/gapminderDataFiveYear (1).csv")
```

```python
countries = ['France', 'United Kingdom', 'Italy', 'Germany', 'Spain']

europe_df = df[df['country'].isin(countries)]
```

```python
plt.figure(figsize=(8, 5))

for country in countries:
    country_data = europe_df[europe_df['country'] == country]
    plt.plot(
        country_data['year'],
        country_data['pop'],
        marker='o',
        linewidth=2,
        label=country
    )
```

#####diamonds.csv file contains various attributes like carat, cut, color, clarity, price etc. for over 53,000 diamonds. Visualize the relationship between the carat (size of diamond) and price using a scatter plot. Instead of using the entire dataset for this visualization, just pick the diamonds with a clarify "SI2" and color "E". Use the values of the "cut" column to color the dots in the scatter plot. Make appropriate modifications to the chart title, axis titles, legend, figure size, font size, colors etc. to make the chart readable and visually appealing. -# Scatterplot

```python
df = pd.read_csv("D:/ml/MachineLearning2026_Aditi/Datasets/diamonds (1).csv")
```

```python
filtered_df = df[
    (df['clarity'] == 'SI2') &
    (df['color'] == 'E')
]
```

```python
plt.figure(figsize=(9,7))
sns.scatterplot(
    data=filtered_df,
    x="carat",
    y="price",
    hue="cut",
    palette="viridis",
    alpha=0.7,
    s=70
)

# Chart customizations
plt.title("Carat vs Price of Diamonds (Clarity: SI2, Color: E)", fontsize=16, fontweight="bold")
plt.xlabel("Carat (Diamond Size)", fontsize=13)
plt.ylabel("Price (USD)", fontsize=13)
plt.legend(title="Cut Type", title_fontsize=12, fontsize=11)
plt.tight_layout()

# Show plot
plt.show()
```

#####The Planets dataset contains details about the 1,000+ extrasolar planets discovered up to 2014. Visualize the distribution of the masses of the planets (expressed as a multiple of the mass of Jupiter), using a histogram and a box plot. Make appropriate modifications to the chart title, axis titles, legend, figure size, font size, colors etc. to make the chart readable and visually appealing. - # Histogram and Box Plot

```python
df = pd.read_csv("D:/ml/MachineLearning2026_Aditi/Datasets/planets (1).csv")
```

```python
mass_data = df['mass'].dropna()
```

```python
sns.set(style="whitegrid", font_scale=1.2)
```

```python
plt.figure(figsize=(14, 6))
```

```python
plt.subplot(1, 2, 1)
plt.hist(
    mass_data,
    bins=30,
    color="steelblue",
    edgecolor="black",
    alpha=0.8
)
plt.title("Distribution of Exoplanet Masses", fontsize=16, fontweight="bold")
plt.xlabel("Planet Mass (Jupiter Masses)", fontsize=13)
plt.ylabel("Frequency", fontsize=13)
plt.grid(axis="y", alpha=0.4)
```

```python
plt.subplot(1, 2, 2)
sns.boxplot(
    x=mass_data,
    color="orange",
    width=0.5
)
plt.title("Box Plot of Exoplanet Masses", fontsize=16, fontweight="bold")
plt.xlabel("Planet Mass (Jupiter Masses)", fontsize=13)
plt.tight_layout()
plt.show()
```
